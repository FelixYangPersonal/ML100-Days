2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
def create_data(vmin=-1.0, vmax=1.0, nrow=10, ncol=10, method=1):
    if method == 1:
        data = np.random.uniform(vmin, vmax, size=[nrow, ncol])
    if method == 2:
        data = np.random.normal(vmin, vmax, size=[nrow, ncol])
    return data
```

```python
vmin = -1.0
vmax = 1.0

matrix = create_data(vmin, vmax)

plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix, vmin=vmin, vmax=vmax, cmap=plt.cm.RdYlBu_r, annot=True)

plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = create_data(vmin, vmax, nrow, ncol)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

grid = sns.PairGrid(data=plot_data, size=3, diag_sharey=False)

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = create_data(nrow=nrow, ncol=ncol, method=2)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
